```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("belgian_property_prices.csv")
```

```python
df.columns
```

```python
df1 = df[["price", "living_area", "garden_surface", "energy_class", "covered_parking_spaces", "outdoor_parking_spaces", "construction_year", "surroundings_type"]]
```

```python
df1.head()
```

```python
df1 = df1[df1['surroundings_type'] == 'Urban']
```

```python
#filling NA values for parking space and garden surface with 0

df1["covered_parking_spaces"] = df1["covered_parking_spaces"].fillna(0)


df1["garden_surface"] = df1["garden_surface"].fillna(0)

df1["outdoor_parking_spaces"] = df1["outdoor_parking_spaces"].fillna(0)
```
